```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split #splitting the dataset
from autogluon.tabular import TabularDataset, TabularPredictor #to handle tabular data and train models
```

```python
import numpy as np
#import dataset of QSAR features and target value
df_bio=pd.read_csv("biodeg.csv", sep = ';', header=None)
```

```python
#Assign values for target labels of features
d = {'RB': 1, 'NRB': 0}

df_bio['bio_class']=df_bio.iloc[:,-1].map(d)
df_bio.drop(df_bio.columns[41],axis=1,inplace=True)
```

```python
df_bio.head()
```

```python
# split into train and test sets
df_train,df_test=train_test_split(df_bio,test_size=0.33,random_state=1)
df_train.shape,df_test.shape
```

```python
# Check the Features and new target column type
df_train.info()
```

```python
test_data=df_test.drop(['bio_class'],axis=1)
```

```python
#Use the default arguments on the first pass before fine tuning it
predictor= TabularPredictor(label ='bio_class').fit(train_data = df_train, verbosity = 2,presets='best_quality')
```

```python
#Get the summary of the models for the training scores
predictor.fit_summary()
```

```python
predictor.leaderboard(df_train, silent=True)
```

```python
#Evaluate it with the test data and convert the predictions to a dataframe
y_pred = predictor.predict(test_data)
y_pred=pd.DataFrame(y_pred,columns=['bio_class'])
```

```python
predictor.evaluate(df_test)
```
